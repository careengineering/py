Random Modülü

```python
import random
random.__dir__()
```

```python
help(random)
```

```python
random.random() # 0 ile 1 arasında ondalık sayı bulur
```

```python
random.uniform(5,10) # 5-10 arasında ranstgele bi sayı
```

```python
random.randint(5,10) #rastgele int
```

```python
random.choice([*range(10)]) # verilen aralıkta rastgele seçim yapar
```

```python
random.sample(range(10), k=4) #listeden rastgele 4 sayı seçer
```

```python
liste = [*range(10)]
print(liste)
random.shuffle(liste) #listeyi karıştırma. sonucu yokluk olduğu için değişkene atama
print(liste)
```
