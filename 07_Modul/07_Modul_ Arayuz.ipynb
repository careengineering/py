```python
from tkinter import * #arayüz için en sık kullanılan paket
from tkcalendar import DateEntry #kullanıcı tarafından geliştirilen paketler gömülü değil. terminalde pip install tkcalendar yap


master = Tk() #arayüz oluşur kapanır

#ana ekranı canvasa eklemlendirmek için kullanılan yöntemler aşağıda gösterilmektedir.
#pack
#place
#grid

canvas = Canvas(master,height=450,width=750)
canvas.pack()

frame_ust = Frame(master,bg="#ADD8E6")
frame_ust.place(relx=0.1,rely=0.1,relwidth=0.8,relheight=0.1) #relative başlangıç noktasından uzaklık gösterir. %80 0.8 oluyor

frame_alt_sol = Frame(master,bg="#ADD8E6")
frame_alt_sol .place(relx=0.1,rely=0.21,relwidth=0.23,relheight=0.5)

frame_alt_sag = Frame(master,bg="#ADD8E6")
frame_alt_sag .place(relx=0.34,rely=0.21,relwidth=0.56,relheight=0.5)

hatirlatma_tipi_etiket = Label(frame_ust,bg="#ADD8E6", text="Hatirlatma tipi:",font="Verdana 12 bold")
hatirlatma_tipi_etiket .pack(padx=10, pady=10, side=LEFT) #pad kenara mesafe , side nereye yaslanacağı

hatirlatma_tipi_opsiyon = StringVar(frame_ust) #string değişken tanımlandı
hatirlatma_tipi_opsiyon.set("\t") #default değer atama
hatirlatma_tipi_acilir_menu = OptionMenu(
    frame_ust, #hangi ebeveyn içinde
    hatirlatma_tipi_opsiyon, #default değer
    "Dogum gunu",
    "Alisveris",
    "Odeme")
hatirlatma_tipi_acilir_menu.pack(padx=10, pady=1, side=LEFT)

hatirlatma_tarih_secici = DateEntry(frame_ust, width=12, background="orange", foreground="black", borderwidth=1, locale="de_DE" )
hatirlatma_tarih_secici._top_cal.overrideredirect(False)
hatirlatma_tarih_secici.pack(padx=10, pady=10, side=RIGHT)


hatirlatma_tarihi_etiket = Label(frame_ust,bg="#ADD8E6", text="Hatirlatma tarihi:",font="Verdana 12 bold")
hatirlatma_tarihi_etiket .pack(padx=10, pady=10, side=LEFT) #pad kenara mesafe , side nereye yaslanacağı

#herşeye değişken atamak zorunda değiliz. Memory yer kaplamamak için
#statik değerleri değişkensiz tanımlayabiliriz
Label(frame_alt_sol,bg="#ADD8E6", text="Hatirlatma yontemi:",font="Verdana 12 bold").pack(padx=10, pady=10, anchor=NW) 

var = IntVar() #radiobuttonda var değişkeni değeri değişecek
R1 = Radiobutton(frame_alt_sol,text="Sisteme Kaydet", variable=var, value=1, bg="#ADD8E6",font="Verdana 12 bold" )
R1.pack(anchor=NW, pady=5, padx=15)

R2 = Radiobutton(frame_alt_sol,text="Eposta Gönder", variable=var, value=2, bg="#ADD8E6",font="Verdana 12 bold" )
R2.pack(anchor=NW, pady=5, padx=15)

var1=IntVar()
C1=Checkbutton(frame_alt_sol,text="Bir hafta önce", variable=var1, onvalue=1, offvalue=0, bg="#ADD8E6",font="Verdana 12 bold" )
C1.pack(anchor=NW, pady=2, padx=25)

var2=IntVar()
C2=Checkbutton(frame_alt_sol,text="Bir gün önce", variable=var2, onvalue=1, offvalue=0, bg="#ADD8E6",font="Verdana 12 bold" )
C2.pack(anchor=NW, pady=2, padx=25)

var3=IntVar()
C3=Checkbutton(frame_alt_sol,text="Aynı gün", variable=var3, onvalue=1, offvalue=0, bg="#ADD8E6",font="Verdana 12 bold" )
C3.pack(anchor=NW, pady=2, padx=25)

from tkinter import messagebox
def gonder():
    son_mesaj=""
    try:
        if var.get():
            if var.get()==1:
                son_mesaj += "Veriniz başarıyla sisteme kaydedilmiştir."
                tip = hatirlatma_tipi_opsiyon.get() if hatirlatma_tipi_opsiyon.get() == "" else "Genel"
                tarih = hatirlatma_tarih_secici.get()
                mesaj = metin_alani.get("1.0","end" )#get komutu text te çalışmıyor

                with open("hatirlatmalar.txt", "w") as dosya:
                    dosya.write(
                        "{} kategorisinde, {} tarihine '{}' notuyla hatırlatma".format(
                            tip,
                            tarih,
                            mesaj
                        ))
                    dosya.close()
            elif var.get()==2:
                son_mesaj += "Eposta yoluyla hatırlatma size ulaşacaktır."
            
            messagebox.showinfo("Başarılı işlem", son_mesaj)
        else:
            son_mesaj += "Gerekli alanları doldurulduğundan emin olun!"
            messagebox.showwarning("Yetersiz bilgi", son_mesaj)
    except:
        son_mesaj += "İşlem başarısız oldu!"
        messagebox.showerror("Başarısız işlem", son_mesaj)

    finally:
        master.destroy()


Label(frame_alt_sag,bg="#ADD8E6", text="Hatirlatma mesajı:",font="Verdana 12 bold").pack(padx=10, pady=10, anchor=NW) 

metin_alani = Text(frame_alt_sag, height=9, width=50)
metin_alani.tag_configure("style",foreground="#bfbfbf", font=("Verdana",7,"bold"))
metin_alani.pack()
karsilama_metni ="Mesajı buraya gir..."
metin_alani.insert(END, karsilama_metni,"style")

gonder_butonu = Button(frame_alt_sag, text="Gönder", command=gonder) #command=master.destroy yazarak master öğesini kapatır
gonder_butonu.pack(anchor=S)

master.mainloop() #arayüzün açık kalması için
```
